# Stochastic simulation with `stochpy`

- [doc](http://stochpy.sourceforge.net/html/userguide_doc.html)
- [repo](https://github.com/SystemsBioinformatics/stochpy)

```python
%matplotlib inline
import matplotlib.pyplot as plt
import stochpy
import warnings
```

```python
%%file /Users/cliburnchan/Stochpy/pscmodels/birth_death.psc

# Reactions
R1:
    $pool > x
    lam*x

R2:
    x > $pool
    mu*x
    
# Fixed species
 
# Variable species
x = 50
 
# Parameters
lam = 3
mu = 4
```

```python
smod = stochpy.SSA()
smod.Model('birth_death.psc')
smod.DoStochSim(end = 5, mode = 'time', trajectories = 3)
smod.PlotSpeciesTimeSeries()
smod.PlotAverageSpeciesTimeSeries(linestyle='solid')
smod.PlotAverageSpeciesDistributions()
```

```python
smod.DoStochSim(end = 5, mode = 'time', trajectories = 3, method='NextReactionMethod')
smod.PlotSpeciesTimeSeries()
```

```python
smod.DoStochSim(end = 5, mode = 'time', trajectories = 3, method='FirstReactionMethod')
smod.PlotSpeciesTimeSeries()
```

```python
with warnings.catch_warnings():
    warnings.simplefilter('ignore', RuntimeWarning)
    smod.DoStochSim(end = 5, mode = 'time', trajectories = 3, method='TauLeaping')
smod.PlotSpeciesTimeSeries()
```
